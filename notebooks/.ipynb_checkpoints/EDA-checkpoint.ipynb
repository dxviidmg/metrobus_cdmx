# Obtención de alcaldia por medio de geo points

```python
def format_geographic_point(**kwargs):
    """
    Formateo del punto geografico
    """
    if 'position_latitude' in kwargs.keys() and 'position_longitude' in kwargs.keys():
        return str(kwargs['position_latitude']) + ', ' + str(kwargs['position_longitude'])
        
    return 'Error, formato no definido'

def get_location(data):
    
    if isinstance(data, str):
        geographic_point = data
    else:
        geographic_point = format_geographic_point(**data)
    
    geolocator = Nominatim(user_agent="http")
    location = geolocator.reverse(geographic_point)
    if location == None:
        return None
    
    location = location
    return location
```

```python
# Obtencion de todos los geopoints
df['location'] = df[['position_latitude', 'position_longitude']].apply(get_location, axis=1)
df
```

## Obtención de alcaldias 

```python
def get_alcaldia(data):
    location = get_location(data)
    if location == None:
        return 'N/A'
    
    location = list(location)[0]
        
#    print(location)
    location_list = location.split(', ')
    
    if 'Ciudad de México' in location_list:
        entidad = 'Ciudad de México'
    elif 'Estado de México' in location_list:
        entidad = 'Estado de México'
        
    index_referencia = location_list.index(entidad)
    
    index_referencia = index_referencia - 1
#    print(index_referencia)
    alcaldia = location_list[index_referencia]
    print(alcaldia)
    return alcaldia
```

```python
# Ejemplo de alcaldia por medio de geopoints
get_alcaldia('19.4792003631592, -99.0639038085938')
```

```python
#alcaldias del df
df['alcaldia'] = df[['position_latitude', 'position_longitude']].apply(get_alcaldia, axis=1)
df
```

```python
alcaldias = df.alcaldia.unique()
alcaldias.sort()
alcaldias
```
